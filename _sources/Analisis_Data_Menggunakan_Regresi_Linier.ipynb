# Analisis Data Menggunakan Regristrasi Linier #

- Proyek ini digunakan untuk membuat analisis data menggunakan metode Regresi Linier. Analisis dilakukan untuk mengetahui hubungan antara variabel x dan y serta mencari persamaan garis regresi. Perhitungan dilakukan menggunakan aplikasi GeoGebra dan library sklearn pada Python.





# 1. Data Penelitian

Data yang digunakan adalah sebagai berikut:

| Titik | x | y |
| ----- | - | - |
| A   =  |( 2 |, 2 )|
| B   =  |( 4 |, 3 )|
| C   =  | (5 |, 5 )|
| D  =   |( 3 |, 4 )|
| E   =  |( 3 |, 3 )|
| F   =  |( 4 |, 5 )|
| G   =  |( 5 |, 6 )|


# 2. Langkah Pengerjaan di GeoGebra

**- Memasukkan Data Titik**

Data dimasukkan ke GeoGebra menggunakan format titik koordinat:

```python
A=(2,2)
B=(4,3)
C=(5,5)
D=(3,4)
E=(3,3)
F=(4,5)
G=(5,6)
```

**- Membuat List Data**

Seluruh titik digabungkan menjadi satu list menggunakan perintah:

**L={A,B,C,D,E,F,G}**

**-Membuat Garis Regresi**

Garis regresi dibuat menggunakan perintah:

**FitLine(L)**

GeoGebra kemudian menampilkan garis regresi beserta persamaannya.
