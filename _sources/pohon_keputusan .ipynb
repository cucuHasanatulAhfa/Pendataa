# 2. Preprocessing Data

Tahapan preprocessing data dilakukan sebagai langkah awal untuk menyiapkan dataset sebelum digunakan dalam proses pemodelan. Pada tahap ini, data akan melalui proses pembersihan, seleksi, serta transformasi ke dalam bentuk yang lebih terstruktur sehingga dapat diolah secara optimal oleh algoritma machine learning.

Adapun langkah-langkah preprocessing yang dilakukan adalah sebagai berikut:



  # a. CSV Reader

  Node CSV Reader digunakan untuk mengimpor dataset diabetes yang tersimpan dalam format file .csv. Setelah proses pembacaan selesai, data akan ditampilkan dalam bentuk tabel yang memuat seluruh atribut (fitur) beserta kelas targetnya.


  # b.Table Partition

  Screenshot (1251).png


  Node Table Partitioner digunakan untuk memisahkan dataset menjadi dua bagian utama, yaitu data latih (training) dan data uji (testing). Data training dimanfaatkan untuk membangun serta melatih model, sedangkan data testing digunakan untuk mengevaluasi kinerja model terhadap data yang belum pernah dilihat sebelumnya.

Proses pembagian ini penting agar hasil evaluasi model menjadi lebih objektif dan tidak bias.

Adapun pembagian data yang digunakan adalah sebagai berikut:

 - 90% Data Training (13 baris): Digunakan untuk proses pembelajaran model dalam mengenali pola dari dataset.

 - 10% Data Testing (1 baris): Digunakan untuk menguji tingkat akurasi model dalam memprediksi data baru.

# e. Decision Tree Learner

Node Decision Tree Learner berperan sebagai inti pemrosesan yang menganalisis pola dari data training untuk membentuk aturan prediksi pada kolom Play Tennis. Proses pembentukan pohon keputusan ini menggunakan pendekatan statistik Gain Ratio agar pemilihan atribut menjadi lebih optimal.

Penjelasan Hasil Decision Tree
Root Node (Akar Utama): Wind

Berdasarkan hasil pada gambar, atribut pertama yang dipilih sebagai pemisah utama adalah Wind (Angin).
Hal ini menunjukkan bahwa kondisi angin menjadi faktor paling berpengaruh dalam menentukan apakah seseorang akan bermain tenis atau tidak.

Dari total 12 data training, terdapat:

  - 8 data “Yes” (66,7%)
 - 4 data “No” (33,3%)

Percabangan Berdasarkan Wind

Data kemudian dibagi menjadi dua kondisi:

1. Wind = False
  - Total data: 7
   - Yes: 5 (71,4%)
  - No: 2 (28,6%)

Artinya, ketika tidak ada angin, kemungkinan bermain tenis lebih besar (mayoritas memilih Yes).

2. Wind = True
   - Total data: 5
   - Yes: 3 (60%)
   - No: 2 (40%)

Pada kondisi berangin, masih lebih banyak yang memilih Yes, tetapi selisihnya tidak terlalu besar dibandingkan kondisi tanpa angin.

Kesimpulan

Model decision tree menunjukkan bahwa:

 - Faktor Wind menjadi penentu utama dalam pengambilan keputusan.

 - Pada kedua kondisi (True maupun False), hasil tetap didominasi oleh Yes.
 - Namun, peluang bermain tenis lebih tinggi saat tidak ada angin.
