# Mengukur Jarak Campuran
**Data Understanding – Dataset Tips**

Dataset Tips merupakan dataset yang berisi informasi mengenai tagihan restoran dan perilaku pemberian tip pelanggan. Dataset ini sering digunakan dalam analisis data untuk mempelajari hubungan antara tagihan, jumlah tip, jumlah orang, serta karakteristik pelanggan.

Dataset ini terdiri dari:

 - 244 data

 - 7 atribut

Dataset ini memiliki tipe data campuran, yaitu:

 - **Numerik**

 - **Kategorikal**

Sehingga untuk menghitung kemiripan antar data diperlukan metode jarak campuran (mixed distance).

#Transformasi Atribut Sebelum Perhitungan Jarak

Karena dataset memiliki tipe data campuran maka dilakukan transformasi agar semua atribut dapat dihitung secara matematis.

#1. Atribut Numerik

Atribut numerik:

 - total_bill

 - tip

 - size

Dilakukan normalisasi menggunakan Min-Max Scaling

**Rumus:**

$$
x'=\frac{x-\min(x)}{\max(x)-\min(x)}
$$

Tujuan normalisasi adalah menyamakan skala data agar tidak ada atribut yang mendominasi perhitungan jarak.

#2. Atribut Kategorikal (Nominal)

Atribut:

 - sex

 - smoker

 - day

 - time

Ditranformasikan menggunakan label encoding.

Contoh pengkodean:

**Sex**

Male = 0
Female = 1

**Smoker**

No = 0
Yes = 1

**Day**

Thur = 0
Fri = 1
Sat = 2
Sun = 3

**Time**

Lunch = 0
Dinner = 1

Perhitungan jarak atribut kategorikal:

$$
d(x,y)=
\begin{cases}
0, & \text{jika kategori sama} \\
1, & \text{jika kategori berbeda}
\end{cases}
$$

#Perhitungan Jarak Campuran

Rumus umum jarak:

$$
d(i,j)=\sum_{k=1}^{p} d_k(i,j)
$$

dengan:

 - p = jumlah atribut

 - 𝑑
 𝑘= jarak tiap atribut

$$

$$
Hasil perhitungan menghasilkan nilai jarak yang menunjukkan tingkat kemiripan antara dua data pelanggan.

#Analisis Singkat
Diagonal Bernilai Nol

Nilai diagonal pada matriks jarak selalu 0 karena jarak data terhadap dirinya sendiri adalah nol.

Sifat Simetris

Jarak bersifat simetris

$$
d(A,B) = d(B,A)
$$
