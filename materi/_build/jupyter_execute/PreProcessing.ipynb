# Preprocessing

**1. Pengertian Preprocessing**

Preprocessing data adalah tahap awal dalam proses data mining atau machine learning yang dilakukan untuk mempersiapkan data sebelum dianalisis lebih lanjut.

Data yang diperoleh dari berbagai sumber biasanya masih berupa data mentah yang belum siap digunakan secara langsung. Dalam data tersebut sering ditemukan beberapa permasalahan seperti:

 - Data yang hilang (missing value)

 - Data tidak konsisten

 - Skala data yang berbeda antar atribut

 - Data yang mengandung kesalahan atau noise

Oleh karena itu, preprocessing dilakukan untuk membersihkan dan menyiapkan data agar lebih terstruktur sehingga proses analisis dapat berjalan dengan lebih baik dan menghasilkan model yang lebih akurat.

# 2. Tahapan Preprocessing Data.

Dalam proses preprocessing terdapat beberapa tahapan yang umum dilakukan, yaitu:

 **1. Data Cleaning**

Tahap ini bertujuan untuk membersihkan data dari kesalahan, nilai yang tidak valid, atau data yang hilang.

**2. Data Integration**

Proses penggabungan data yang berasal dari beberapa sumber sehingga menjadi satu dataset yang lengkap.

**3. Data Transformation**

Tahap mengubah bentuk atau format data agar sesuai dengan kebutuhan analisis.

**4. Data Normalization**

Proses mengubah nilai data ke dalam skala yang sama sehingga tidak ada atribut yang memiliki pengaruh lebih besar hanya karena nilainya lebih besar.

# 3. Normalisasi Data

Normalisasi data merupakan proses mengubah nilai suatu atribut ke dalam skala tertentu tanpa mengubah hubungan antar data tersebut.

Tujuan dari normalisasi data antara lain:

 - Menyamakan rentang nilai antar atribut

 - Menghindari dominasi nilai yang besar

 - Mempermudah proses perhitungan pada algoritma machine learning seperti KNN, WKNN, dan clustering

Beberapa metode normalisasi yang sering digunakan yaitu:

 - Min-Max Normalization

 - Z-Score Normalization

 - Decimal Scaling Normalization


 # 3.1 Min-Max Normalization

Min-Max Normalization adalah metode normalisasi yang mengubah nilai data ke dalam rentang 0 sampai 1.

Rumus Min-Max:

$$
X_{norm} = \frac{X - X_{min}}{X_{max} - X_{min}}
$$

Keterangan:

X = nilai data asli

Xmin = nilai terkecil pada data

Xmax = nilai terbesar pada data

Xnorm = nilai hasil normalisasi

 Perhitungan

Terdapat data IPK dari dataset pada tabel:

2, 3, 4

Nilai minimum = 2
Nilai maksimum = 4

Jika ingin menormalisasi nilai 3:

(3 - 2)/(4 - 2) = 0.5

Hasil normalisasi:

| Nilai Asli | Hasil Normalisasi |
| ---------- | ----------------- |
| 2          | 0                 |
| 3          | 0.5               |
| 4          | 1                 |

# 3.2 Z-Score Normalization

Z-Score Normalization adalah metode normalisasi yang menggunakan nilai rata-rata (mean) dan standar deviasi untuk mengubah data ke dalam bentuk distribusi standar.

Rumus Z-Score:

$$
Z = \frac{X - \mu}{\sigma}
$$

 Keterangan:

X = nilai data

μ = rata-rata data

σ = standar deviasi

Z = nilai normalisasi

 Perhitungan

Data IPK:

2, 3, 4

Mean:

  (2 + 3 + 4) /3=3

Standar deviasi dihitung dari selisih tiap data dengan rata-rata.

Jika ingin menormalisasi nilai 3:

$$
Z = \frac{3 - 3}{\sigma} = 0
$$

Artinya nilai tersebut berada tepat pada rata-rata data.


# 3.3 Decimal Scaling Normalization

Decimal Scaling Normalization dilakukan dengan cara membagi nilai data dengan pangkat 10 berdasarkan jumlah digit terbesar pada data.

Rumus:

$$
X_{norm} = 10X
$$

Keterangan:

X = nilai data

j = jumlah digit terbesar pada data

**Perhitungan**

Data PO pada tabel:

200000

300000

400000

Nilai terbesar = 400000

Jumlah digit = 6

Maka normalisasi dilakukan dengan membagi 10⁶.

Contoh:

200000 / 1000000 = 0.2

300000 / 1000000 = 0.3

400000 / 1000000 = 0.4

Hasil normalisasi:

| Nilai Asli | Hasil Normalisasi |
| ---------- | ----------------- |
| 200000     | 0.2               |
| 300000     | 0.3               |
| 400000     | 0.4               |


# 4. Kesimpulan

Preprocessing data merupakan tahap penting dalam pengolahan data sebelum dilakukan analisis atau pemodelan machine learning. Salah satu proses penting dalam preprocessing adalah normalisasi data yang bertujuan untuk menyamakan skala antar atribut.

Beberapa metode normalisasi yang sering digunakan antara lain:

 - Min-Max Normalization untuk mengubah data ke rentang 0 sampai 1

 - Z-Score Normalization menggunakan nilai rata-rata dan standar deviasi

 - Decimal Scaling Normalization dengan membagi data menggunakan pangkat 10

Dengan melakukan normalisasi, proses analisis data menjadi lebih stabil dan hasil perhitungan algoritma menjadi lebih akurat.
