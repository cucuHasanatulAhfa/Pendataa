 **- Impute **

Fungsi:

Mengatasi missing value (data kosong) pada dataset.

Pengaturan yang dipakai:
 - Average/Most frequent
   - Data numerik → diisi rata-rata
   - Data kategorikal → diisi modus

   jadi tidak ada data kosong dan data jadi siap di prosses ke tahap berikutnya

**Continuize**

Fungsi:

Mengubah data kategorikal menjadi numerik

Yang diproses:

 - Tekstur Tanah (kategorikal) → jadi angka

Pengaturan:
- Menggunakan First value as base / encoding

jadi KNN butuh data numerik dan semua data jadi bisa di hitung.

impute -> membersihkan data dari nilai kosong .

continuize -> mengubah data biar dipakai KNN

Screenshot (1161).png

 **Normalize Features :**

Digunakan untuk menyamakan skala data agar tiap fitur punya pengaruh seimbang.

Pilihan:


- Standardize to μ=0, σ=1 → Mengubah data jadi rata-rata 0 dan standar deviasi 1 (paling umum dipakai).


- Center to μ=0 → Menggeser data supaya rata-rata = 0.


- Scale to σ=1 → Mengatur agar standar deviasi = 1.


- Normalize to [-1, 1] → Mengubah nilai ke rentang -1 sampai 1.


- Normalize to [0, 1] → Mengubah nilai ke rentang 0 sampai 1 (min-max).

jadi biar data tidak beda skala (jumlah haarga dan jumlah )

**k-NN**

Digunakan untuk klasifikasi berdasarkan tetangga terdekat.

 - Number of neighbors (k = 5) → jumlah tetangga yang dipakai untuk menentukan kelas.

- Metric: Euclidean → cara menghitung jarak antar data.

- Weight: Uniform → semua tetangga punya pengaruh sama.

jadi data baru diklasifikasikan berdasarkan mayoritas tetangga terdekat.

Screenshot (1163).png

**Test & Score**

Digunakan untuk mengukur performa model.

- Cross Validation (5 folds) → data dibagi 5 bagian untuk uji akurasi.

- AUC = 1.000 → model sangat baik membedakan kelas.

- CA (Accuracy) = 1.000 → akurasi 100%.

- F1 = 1.000 → keseimbangan precision & recall sempurna.

- Precision = 1.000 → prediksi benar semua.

- Recall = 1.000 → semua data berhasil dikenali.

- MCC = 1.000 → korelasi prediksi sangat sempurna.

 nilai 1.000 menunjukkkan performa sangat tinggi, namum perlu dicek kemungkinan overfitting.

**Confusion Matrix**

Menunjukkan hasil prediksi vs data asli.

- Subur → Subur = 1000 (benar)

-  Tidak Subur → Tidak Subur = 1000 (benar)

- Salah prediksi = 0

fungsinya Untuk mengevaluasi kinerja model dengan melihat jumlah prediksi yang benar dan salah pada tiap kelas.

juga untuk membantu mengetahui seberapa akurat model dan apakah ada kesalahan klasifikasi.

**Scatter Plot**

Menampilkan hubungan antar variabel:

- Sumbu X: pH Tanah

 - Sumbu Y: C Organik (%)

- Warna:

- Biru = Subur

- Merah = Tidak Subur

Scatter plot digunakan untuk melihat pola hubungan antara pH tanah dan kandungan C Organik serta bagaimana kedua variabel tersebut mempengaruhi klasifikasi kesuburan tanah. Dari grafik terlihat bahwa data kelas Subur dan Tidak Subur membentuk kelompok yang terpisah dengan jelas tanpa banyak tumpang tindih.
