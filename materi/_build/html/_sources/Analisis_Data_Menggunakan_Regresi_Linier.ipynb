## Menghitung Koefisien Regresi

Rumus koefisien regresi:

$$
b = \frac{n\sum xy - (\sum x)(\sum y)}
{n\sum x^2 - (\sum x)^2}
$$

Substitusi nilai:

$$
b = \frac{7(112) - (26)(28)}
{7(104) - (26)^2}
$$

$$
b = \frac{784 - 728}
{728 - 676}
$$

$$
b = \frac{56}{52}
$$

$$
b \approx 1.077
$$

## Menghitung Konstanta

Rumus konstanta:

$$
a = \frac{\sum y - b\sum x}{n}
$$

Substitusi nilai:

$$
a = \frac{28 - (1.077)(26)}{7}
$$

$$
a = \frac{28 - 28.002}{7}
$$

$$
a = \frac{-0.002}{7}
$$

$$
a \approx -0.0003 \approx 0
$$

## Persamaan Regresi

$$
y = a + bx
$$

$$
y = 0 + 1.077x
$$

$$
y = 1.077x
$$

# 4. Persamaan Regresi

Persamaan regresi yang diperoleh adalah:

y = 1.077x


# 5.Implementasi Menggunakan Python dan Sklearn #

Kode program menggunakan library sklearn:

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# data
x = np.array([2,4,5,3,3,4,5]).reshape(-1,1)
y = np.array([2,3,5,4,3,5,6])

# model regresi
model = LinearRegression()
model.fit(x, y)

# hasil
print("Koefisien :", model.coef_[0])
print("Intercept :", model.intercept_)
```

# 6. Hasil Program #

Hasil dari program menunjukkan:

 - Koefisien regresi = 1.077

 - Intercept ≈ 0

Sehingga persamaan regresi yang dihasilkan adalah:

y = 1.077x
